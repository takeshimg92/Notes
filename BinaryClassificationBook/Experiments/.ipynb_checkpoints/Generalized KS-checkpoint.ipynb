# Generalized KS score

Let $F, G$ be the CDFs of two real random variables. **Assume** (and this is not obvious) that both $F, G$ have compact support, ie. there exist numbers $a,b$ with $b>a$ such that $F$ and $G$ are zero outside $[a,b]$. Let then $[a,b]$ be the smallest interval satisfying this requirement.

Denote, for any function $f$,

$$\|f\|_p :=\left( \int_a^b |f(x)|^p \; dx\right)^{1/p}$$

as the $L^p([a,b])$ norm of $f$. It is easy to show that

$$ \|F-G\|_p \leq (b-a)^{1/p} \max{x\in[a,b]}|F(x) - G(x)|,$$

or equivalently

$$\frac{1}{(b-a)^{1/p}} \|F-G\|_p  \leq \max{x\in[a,b]}|F(x) - G(x)| \equiv \mathrm{KS}(F,G)$$

where we identify the RHS with the KS statistic over $[a,b]$.

This motivates the definition of

$$\boxed{D_p(F, G) = \frac{1}{(b-a)^{1/p}} \|F-G\|_p }$$

as a divergence between $F$ and $G$. This quantity, for fixed $p$, is clearly a distance (since it inherits the distance properties from the $L^p$ norm), and as $p\to \infty$, the equality is attained.

```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']

plt.style.use('dark_background')
import warnings
warnings.filterwarnings('ignore', '.*If increasing the limit yields no improvement .*')
```

## Obtendo distribuicoes acumuladas a partir de uma serie de dados

```python
def d_p(F, G, p, a, b):
    
    """
    Calculates the Lp norm (divided by (b-a)**(1/p)) as a distance metric between
    CDFs F and G, which are supposed to be function objects (obtained eg. via interpolation)
    """
    
    import scipy.integrate as integrate
    from scipy.optimize import fmin
    
    if p == np.infty:

        max_x = fmin(lambda x: -np.abs(F(x) - G(x)), a, b, full_output=True, disp=False)
        
        return -(max_x[1])
    else:

        result = integrate.quad(lambda x: np.abs(F(x)  - G(x))**p, a, b)

        integral = result[0]**(1/p)

        return integral /(b-a)**(1/p)
```

```python
def cdf(x: np.array):
    """
    Approximate a random variable's CDF via linear interpolation
    of the empirical CDF points
    """
    from scipy.interpolate import interp1d
    
    y = np.arange(len(x))
    y = y/y.max()
    
    order_x = np.sort(x)
    
    f = interp1d(order_x,y, bounds_error=False, fill_value=1)
    a, b = order_x[0], order_x[1]

    def interpolated(p):
        return np.where(p <= a, 0, f(p))

    return interpolated
```

## Criando dois datasets próximos

```python
# generate data
np.random.seed(1)
x1 = np.random.randn(1000)
x2 = 0.5 + np.random.randn(500)

f = cdf(x1)
g = cdf(x2)

# calculate points
p_list = range(1,32, 2)
d_list = [d_p(f, g, p=p, 
    a=min(x1.min(), x2.min()), 
    b=max(x1.max(), x2.max())
   ) for p in p_list]
```

```python
x = np.linspace(-10,10)
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(8,4))

# calculate KS directly
KS = d_p(f, g, 
         p=np.infty, a=min(x1.min(), x2.min()), 
         b=max(x1.max(), x2.max()))

ax[0].plot(x, f(x))
ax[0].plot(x, g(x))
ax[0].legend(['x1', 'x2'])
ax[0].set_title("CDF plot")


ax[1].plot(p_list, d_list)
ax[1].set_xlabel("Power $p$")
ax[1].set_ylabel("Distance $D_p$")
ax[1].axhline(KS, linestyle='--', color='gray')

plt.tight_layout()
plt.show()
```
